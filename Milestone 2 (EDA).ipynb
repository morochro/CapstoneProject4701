```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Capstone1-cleaned-dataset.csv')
```

```python
display(df.head(10))
```

```python
display(df.describe())
```

```python
display(df.info())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
numerical_cols = df.select_dtypes(include=['number']).columns
df[numerical_cols] = df[numerical_cols].apply(lambda x: x.fillna(x.median()))
```

```python
display(df.dtypes)
```

```python
# Missing values check
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values[missing_values > 0])

# Unique values per column
unique_counts = df.nunique()
print("\nUnique values in each column:\n", unique_counts)
```

There are no significant new missing values in the cleaned dataset. The unique value check also helped identify columns with a single unique value that may be excluded in further analysis.

# Subplot

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df['Age'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Age')
sns.histplot(df['BMI'], bins=30, kde=True, ax=axes[1])
axes[1].set_title('Distribution of BMI')
sns.histplot(df['Height'], bins=30, kde=True, ax=axes[2])
axes[2].set_title('Distribution of Height')
plt.show()
```

# Countplot

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 5))

# Countplot for Sex
sns.countplot(x=df['Sex'], ax=axes[0])
axes[0].set_title('Gender Distribution')

# Countplot for Management type
sns.countplot(x=df['Management'], ax=axes[1])
axes[1].set_title('Management Type Distribution')

# Countplot for Severity levels
sns.countplot(x=df['Severity'], ax=axes[2])
axes[2].set_title('Severity Distribution')

# Diagnosis distribution
sns.countplot(x=df['Diagnosis'], ax=axes[3])
axes[3].set_title('Diagnosis Distribution')

plt.tight_layout()
plt.show()
```

# Scatterplot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['WBC_Count'], y=df['CRP'])
plt.title('WBC Count vs CRP')
plt.xlabel('WBC Count')
plt.ylabel('CRP Level')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(x='Diagnosis', y='Age', data=df, ax=axes[0])
axes[0].set_title('Age by Diagnosis')

sns.boxplot(x='Diagnosis', y='WBC_Count', data=df, ax=axes[1])
axes[1].set_title('WBC Count by Diagnosis')

sns.boxplot(x='Diagnosis', y='CRP', data=df, ax=axes[2])
axes[2].set_title('CRP by Diagnosis')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='Sex', hue='Diagnosis', data=df)
plt.title('Sex vs Diagnosis')
plt.show()
```

# Pairplot

```python
sns.pairplot(df[['Age', 'BMI', 'WBC_Count', 'CRP']])
plt.suptitle('Pairplot of Age, BMI, WBC Count, and CRP', y=1.02)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Boxplot of BMI
sns.boxplot(y=df['BMI'], ax=axes[0])
axes[0].set_title('Boxplot of BMI')

# Boxplot of WBC Count
sns.boxplot(y=df['WBC_Count'], ax=axes[1])
axes[1].set_title('Boxplot of WBC Count')

# Boxplot of CRP
sns.boxplot(y=df['CRP'], ax=axes[2])
axes[2].set_title('Boxplot of CRP')

plt.tight_layout()
plt.show()
```

Patients with positive appendicitis diagnoses tend to have higher WBC counts and CRP levels, indicating inflammation. Age and gender do not show strong separation visually but may still contribute when combined with other features.

# Correlation Matrix

```python

plt.figure(figsize=(16, 10))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

From the heatmap, we can observe high correlations between variables such as WBC Count and Neutrophil Percentage, and between BMI and Weight. These strong relationships should be considered during modeling to avoid multicollinearity.

- Most patients fall within a specific age and BMI range, with a normal distribution.
- CRP and WBC Count show strong positive correlation, reflecting inflammation markers.
- Outliers exist in CRP and WBC but may represent real cases of severe inflammation.
- Diagnosis analysis shows that patients diagnosed with appendicitis tend to have higher WBC and CRP levels.
- No major missing values remain, and several categorical features (e.g., Management, Sex) are well-suited for classification models.
